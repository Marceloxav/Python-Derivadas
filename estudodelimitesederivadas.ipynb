```python
from sympy import *
from IPython.display import display
from sympy.plotting import plot

x = symbols('x')
```

```python
f = (x**3 + x**(1/4))/exp(x)

display(f)

L = limit(f, x, oo)

print("Resultado:")
display(L)
```

```python
f = exp(cos(x))*sin(x)

display(f)

D = diff(f, x)

print("Resultado:")
display(D)
```

```python
f = 3*sin(x) - 2*cos(x)

display(f)

F = integrate(f, x)

print("Resultado:")
display(F)
```

```python
plot(f, (x, 0, 2*pi))
```
